```python
import pandas as pd
import altair as alt

# Przykładowe dane wejściowe
df = pd.DataFrame({
    'miesiąc': list(range(1, 13)),
    'plan': [100, 120, 110, 130, 140, 135, 150, 145, 160, 155, 165, 170],
    'wykonanie': [90, 115, 105, 120, 130, 125, 145, 140, 155, 150, 160, 165]
})

# Przekształcenie do formatu długiego
df_melted = df.melt(id_vars='miesiąc', value_vars=['plan', 'wykonanie'],
                    var_name='Typ', value_name='Wartość')

# Konwersja miesiąca na tekst (Altair traktuje oś X jako nominalną)
df_melted['miesiąc'] = df_melted['miesiąc'].astype(str)

# Ustalenie porządku etykiet osi X od "1" do "12"
order_miesięcy = [str(m) for m in range(1, 13)]

# Wykres słupkowy
chart = alt.Chart(df_melted).mark_bar().encode(
    x=alt.X('miesiąc:N', title='Miesiąc', sort=order_miesięcy),
    y=alt.Y('Wartość:Q', title='Wartość'),
    color=alt.Color('Typ:N',
        title='Typ danych',
        scale=alt.Scale(
            domain=['plan', 'wykonanie'],
            range=['#003f5c', '#bfbfbf'] 
        )
    ),
    xOffset='Typ:N'
).properties(
    width=500,
    height=400,
    title='Porównanie planu i wykonania według miesięcy'
)

chart
```

```python
import pandas as pd
import altair as alt
import streamlit as st

# Filtrowanie danych
df_wykres_m = df_plan[df_plan['Lokalizacja'].isin(PLACE)]

# Sumowanie miesięczne
df_wykres_m = df_wykres_m.groupby(['M'], as_index=False)[['P_2025', 'W_2025']].sum()

# Sortowanie po miesiącach (dla poprawnego narastania)
df_wykres_m = df_wykres_m.sort_values('M')

# Obliczanie wartości skumulowanych
df_wykres_m['P_2025_kumul'] = df_wykres_m['P_2025'].cumsum()
df_wykres_m['W_2025_kumul'] = df_wykres_m['W_2025'].cumsum()

# Przekształcenie danych do formatu long
df_melted_kumul = df_wykres_m.melt(
    id_vars='M',
    value_vars=['P_2025_kumul', 'W_2025_kumul'],
    var_name='Typ',
    value_name='Wartość'
)

# Konwersja miesiąca na tekst
df_melted_kumul['M'] = df_melted_kumul['M'].astype(str)

# Mapowanie nazw dla czytelności legendy
df_melted_kumul['Typ'] = df_melted_kumul['Typ'].replace({
    'P_2025_kumul': 'Plan narastająco',
    'W_2025_kumul': 'Wykonanie narastająco'
})

# Kolejność miesięcy
order_M = [str(m) for m in range(1, 13)]

# Wykres liniowy z punktami
line_chart = alt.Chart(df_melted_kumul).mark_line(point=True).encode(
    x=alt.X('M:N', title='Miesiąc', sort=order_M),
    y=alt.Y('Wartość:Q', title='Wartość skumulowana'),
    color=alt.Color('Typ:N', title='Typ danych',
        scale=alt.Scale(
            domain=['Plan narastająco', 'Wykonanie narastająco'],
            range=['#003f5c', '#bfbfbf']
        ),
        legend=alt.Legend(orient='bottom')
    )
).properties(
    width=700,
    height=400,
    title='Plan i wykonanie narastająco – wykres liniowy'
)

line_chart

# Wyświetlenie wykresu w Streamlit
#st.altair_chart(line_chart, use_container_width=True)
```
